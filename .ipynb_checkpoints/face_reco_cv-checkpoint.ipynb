```python
# USAGE
# python detect_faces.py --image rooster.jpg --prototxt deploy.prototxt.txt --model res10_300x300_ssd_iter_140000.caffemodel

# import the necessary packages
import numpy as np
import argparse
import cv2
import time
import matplotlib.pyplot as plt
import os
from PIL import Image
%config Completer.use_jedi = False
confidence_threshold  = 0.9
filepath = "./karm/"
```

```python
#load pretrained model of dnn(deep neural network)
print("[INFO] loading model...")
model = cv2.dnn.readNetFromCaffe("deploy.prototxt.txt", "res10_300x300_ssd_iter_140000.caffemodel")
if model:
    print("model loaded")
```

```python
#load image in numpy array
for filename in ["13_karm.jpg"]:
    image = cv2.imread(filepath+filename)
    im = Image.open(filepath+filename)
    image = np.array(im)
    # cv2.imshow("image",image)
    # cv2.waitKey(3000) 
    # cv2.destroyWindow("image")

    print(image.shape)
    resized_image = cv2.resize(image, (300,300))
    # cv2.imshow("Output",resized_image)
    # time.sleep(5)
    # cv2.destroyWindow("Output")
    #image preprocessing like mean substraction
    blob = cv2.dnn.blobFromImage(resized_image, 1.0,(300, 300), (104.0, 177.0, 123.0))
    print(blob.shape)

    # pass the blob through the network and obtain the detections and
    # predictions
    model.setInput(blob)
    detections = model.forward()
    print(detections.shape)
    image = resized_image
    Image.fromarray(image, 'RGB')
    height, width = image.shape[:2]

    print(detections.shape)
    for each in detections[0][0]:
        confidence = each[2]

        if confidence >0.9 :
            print(confidence)
            startX_factor, startY_factor, endX_factor, endY_factor = each[3:]
            #print(startX_factor)
            startX, startY, endX, endY = int(startX_factor*width), int(startY_factor*height), int(endX_factor*width), int(endY_factor*height)
            print((startX, startY, endX, endY))
            cv2.rectangle(image,(startX,startY),(endX,endY),(0,255,0),2)
            text = str(round(confidence*100,2))
            cv2.putText(image,text,(startX,startY-10),cv2.FONT_HERSHEY_SIMPLEX,0.45,(0,255,0),2)
            print(each[2:])
    Image.fromarray(image, 'RGB').show()
    # cv2.imshow("image",image)
    # cv2.waitKey() 
```
